```python
import pandas as pd
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
print(train_data.shape)
print(test_data.shape)
```

```python
train_data.head()
```

```python
x_train = train_data.drop(['Id','Open Date', 'City', 'City Group', 'Type', 'revenue'  ], axis=1)
x_train.head()
```

```python
y_train = train_data['revenue']
```

```python
x_test = test_data.drop(['Id','Open Date', 'City', 'City Group', 'Type'  ], axis=1)
x_test.head()
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
x_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000)
rf.fit(x_train, y_train)
```

```python
x_test.shape
```

```python
y_test_pred = rf.predict(x_test)
```

```python
x_test['Prediction'] = y_test_pred
```

```python
x_test.head()
```

```python
Final_submission = pd.DataFrame()
Final_submission['Id'] = test_data['Id']
Final_submission['Prediction'] = y_test_pred 
```

```python
Final_submission.head()
```

```python
Final_submission.to_csv('Final_submission.csv', index=False)
```
